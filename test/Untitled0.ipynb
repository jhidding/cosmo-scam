```python
import cairo
from IPython.display import display, SVG, HTML
from math import pi, sqrt
import numpy as np

def Aitoff_Hammer(ra, dec):
    denom = np.sqrt(1 + np.cos(dec) * np.cos(ra/2))
    x = 2 * np.sqrt(2) * np.cos(dec) * np.sin(ra/2) / denom
    y = np.sqrt(2) * np.sin(dec) / denom
    return np.c_[x, y]

class Rot:
    def __init__(self, ra, dec, angle):
        self.p_ra = ra
        self.p_dec = dec
        self.p_angle = angle
        
    def __call__(self, ra, dec):
        v = np.arcsin(np.cos(dec) * np.cos(self.p_dec) * np.cos(ra - self.p_ra) \
                  + np.sin(dec) * np.sin(self.p_dec))
        u = np.arctan2(np.sin(dec) - np.sin(v) * np.sin(self.p_dec),
                       np.cos(dec) * np.sin(ra - self.p_ra) * np.cos(self.p_dec)) + self.p_angle
        return (u+pi)%(2*pi)-pi, v
    
    def inv(self):
        ra,  dec   = self(0, pi/2)
        phi, dummy = Rot(ra, dec, 0)(*self(0, 0))
        return Rot(ra, dec, -phi)
    
    def __mul__(self, other):
        ra,  dec   = other.inv()(self.p_ra, self.p_dec)
        phi, dummy = self(*other(*Rot(ra, dec, 0).inv()(0, 0)))
        return Rot(ra, dec, phi)

eq_to_ga = Rot(np.radians(192.859508), np.radians(27.128336), np.radians(122.932 - 90))
ga_to_eq = eq_to_ga.inv()
ga_to_sg = Rot(np.radians(47.37), np.radians(6.32), 0.0)
sg_to_ga = ga_to_sg.inv()
eq_to_sg = ga_to_sg * eq_to_ga
sg_to_eq = eq_to_sg.inv()

def spherical_to_cartesian(ra, dec, R):
    return np.c_[R * np.cos(dec) * np.cos(ra), \
                 R * np.cos(dec) * np.sin(ra), \
                 R * np.sin(dec)]

from scipy.interpolate import RectBivariateSpline, RectSphereBivariateSpline
from scipy.ndimage import gaussian_filter

def inv_mollweide(l0, x, th):
	return np.arcsin((2*th + np.sin(2*th)) / pi), l0 + pi*x/(2*np.sqrt(2)*np.cos(th))

def fixed_point(f, start):
    n = f(start)
    while np.abs((start - n) / start) > 1e-4:
        start = n
        n = f(start)
    return n

def moll_theta(phi):
    if abs((phi - pi/2)/(pi/2)) < 1e-3:
        return pi/2
    if abs((phi + pi/2)/(pi/2)) < 1e-3:
        return -pi/2
    
    return fixed_point(\
        lambda th: th - (2*th + np.sin(2*th) \
            - pi * np.sin(phi))/(2 + 2 * np.cos(2*th)),
        phi)

class Interp:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __call__(self, X):
        return np.interp(X, self.x, self.y)
    
moll_interp = Interp(np.linspace(-pi, pi, 151), \
    np.array([moll_theta(phi) for phi in np.linspace(-pi, pi, 151)]))

def Mollweide(l, phi, l0=0.0):
    th =  moll_interp(phi)
    return np.c_[2 * np.sqrt(2) / pi * (l - l0) * np.cos(th), \
           np.sqrt(2) * np.sin(th)]
```

```python
def plot_line(cx, proj, l, b):
    path = proj(l, b)
    cx.move_to(*path[0])
    for p in path[1:]:
        x,y = cx.get_current_point()
        if (p[0] - x)**2 + (p[1] - y)**2 < 0.5:
            cx.line_to(*p)
        else:
            cx.move_to(*p)
    cx.stroke()

def prepare_context(cx, N):
    L = 6 
    cx.scale(N/L, N/L)
    cx.translate(3, 1.5)
    cx.set_line_width(0.01)

    cx.save()
    cx.scale(2*sqrt(2), sqrt(2))
    cx.arc(0,0,1,0,2*pi)
    cx.restore()
    cx.set_source_rgb(0.7,0,0)
    cx.stroke()
    
def plot_coords(cx, sph, proj):
    cx.set_source_rgb(0.8, 0.8, 0.8)
    cx.set_line_width(0.008)
    for l in np.linspace(0, 360, 24, endpoint=False):
        b = np.linspace(-90, 90, 50)
        sg_l, sg_b = sph(np.radians(l), np.radians(b))
        plot_line(cx, proj, sg_l, sg_b)

    for b in np.linspace(-75, 75, 11, endpoint=True):
        l = np.linspace(0, 360, 100)
        sg_l, sg_b = sph(np.radians(l), np.radians(b))
        plot_line(cx, proj, sg_l, sg_b)
    
    cx.set_source_rgb(0,0,0.8)
    cx.set_line_width(0.01)
    l_eq = np.linspace(0,360, 100)
    sg_l, sg_b = sph(np.radians(l), np.radians(0))
    plot_line(cx, proj, sg_l, sg_b)

def plot_proj(sph, proj):
    s = cairo.SVGSurface("pynb-test.svg", 600, 300)
    cx = cairo.Context(s)
    prepare_context(cx, 600)
    plot_coords(cx, sph, proj)
    s.finish()
    display(SVG(filename="pynb-test.svg"))
```

```python
sph_id = Rot(0, 0, pi/2)

display(HTML("<H2>Mollweide</H2>"))
plot_proj(eq_to_sg, Mollweide)
plot_proj(ga_to_sg, Mollweide)
plot_proj(sph_id, Mollweide)


display(HTML("<H2>Aitoff-Hammer</H2>"))
plot_proj(eq_to_sg, Aitoff_Hammer)
plot_proj(ga_to_sg, Aitoff_Hammer)
plot_proj(sph_id, Aitoff_Hammer)
```
